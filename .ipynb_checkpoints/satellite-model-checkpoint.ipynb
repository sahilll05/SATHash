```python
import torch
from torch.utils.data import Dataset
import rasterio
import os
import numpy as np

class BigEarthDataset(Dataset):

    def __init__(self, folder):
        self.folder = folder
        self.files = os.listdir(folder)[:5000] # limit dataset

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):

        path = os.path.join(self.folder, self.files[idx])

        with rasterio.open(path) as src:
            img = src.read().astype(np.float32)

        # Sentinel-2 normalization
        img = img / 10000.0

        img = torch.tensor(img)

        return img
```

```python
from torch.utils.data import DataLoader

train_dataset = BigEarthDataset(train_path)

train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True
)

print("Training samples:", len(train_dataset))
```

```python
# FEATURE EXTRACTION FOR RETRIEVAL

import cv2
import numpy as np
import rasterio

def extract_features(path):

    with rasterio.open(path) as src:
        img = src.read()

    img = np.transpose(img,(1,2,0))

    img = cv2.resize(img,(64,64))

    features = img.flatten()

    features = features / np.linalg.norm(features)

    return features
```

```python
# Create Feature Matrix

import os

image_files = []
feature_vectors = []

for f in os.listdir(train_path):

    if f.endswith(".tif"):

        path = os.path.join(train_path,f)

        feat = extract_features(path)

        feature_vectors.append(feat)

        image_files.append(path)

feature_vectors = np.array(feature_vectors)

print("Feature matrix shape:", feature_vectors.shape)
```

```python
# Build FAISS Index

import faiss

dimension = feature_vectors.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(feature_vectors)

print("Total vectors in FAISS:", index.ntotal)
```

```python
# Retrieval Function

def retrieve_similar(query_path, k=5):

    query_feat = extract_features(query_path)
    query_feat = np.expand_dims(query_feat, axis=0)

    D,I = index.search(query_feat,k+1)

    results = []

    for idx in I[0][1:]:   # skip first result
        results.append(image_files[idx])

    return results
```

```python
# Test Retrieval

query = image_files[50]

results = retrieve_similar(query)

print("Query:", query)

for r in results:
    print("Match:", r)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class SpectralHashCNN(nn.Module):

    def __init__(self):

        super(SpectralHashCNN, self).__init__()

        self.conv = nn.Sequential(

            nn.Conv2d(10,32,3,padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),

            nn.Conv2d(32,64,3,padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64,128,3,padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),

            nn.AdaptiveAvgPool2d((1,1))
        )

        # Hash layer
        self.hash = nn.Linear(128,64)


    def forward(self,x):

        x = self.conv(x)

        x = x.view(x.size(0), -1)

        embedding = F.normalize(x, dim=1)

        hash_vector = torch.tanh(self.hash(embedding))

        return embedding, hash_vector
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

model = SpectralHashCNN().to(device)
print(model)
```

```python
# Data Augmentation

import torchvision.transforms as T

augment = T.Compose([
    T.RandomHorizontalFlip(),
    T.RandomVerticalFlip(),
    T.RandomRotation(30)
])
```

```python
# Create Positive Pairs

def create_pairs(batch):

    # original images
    img1 = batch

    # augmented images
    img2 = torch.stack([augment(img) for img in batch])

    # positive pair labels
    labels = torch.ones(batch.size(0), device=batch.device)

    return img1, img2, labels
```

```python
# loss function
import torch
import torch.nn as nn
import torch.nn.functional as F

class ContrastiveLoss(nn.Module):

    def __init__(self, margin=1.0):
        super().__init__()
        self.margin = margin

    def forward(self, emb1, emb2, label):

        distance = F.pairwise_distance(emb1, emb2)

        loss = label * torch.pow(distance, 2) + \
               (1 - label) * torch.pow(torch.clamp(self.margin - distance, min=0.0), 2)

        return torch.mean(loss)


# ------------------------------
# Quantization Loss
# ------------------------------

def quantization_loss(hash_vector):

    return torch.mean((hash_vector.sign() - hash_vector)**2)
```

```python
contrastive_loss = ContrastiveLoss().to(device)
```

```python
# training setup
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

scheduler = torch.optim.lr_scheduler.StepLR(
    optimizer,
    step_size=10,
    gamma=0.5
)
```

```python
# Training Loop

from tqdm import tqdm

epochs = 10

for epoch in range(epochs):

    model.train()

    total_loss = 0

    # tqdm progress bar for batches
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")

    for batch in progress_bar:

        batch = batch.to(device)

        img1, img2, labels = create_pairs(batch)

        img1 = img1.to(device)
        img2 = img2.to(device)

        emb1, hash1 = model(img1)
        emb2, hash2 = model(img2)

        # contrastive loss
        contrastive = contrastive_loss(emb1, emb2, labels)

        # quantization loss
        q_loss = quantization_loss(hash1) + quantization_loss(hash2)

        loss = contrastive + 0.5 * q_loss

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

        # update tqdm progress bar
        progress_bar.set_postfix(loss=loss.item())

    scheduler.step()

    print(f"Epoch {epoch+1}/{epochs} | Total Loss: {total_loss:.4f}")
```

```python
# ================================
# PHASE 11 — HASH GENERATION
# ================================

import numpy as np
import torch
from tqdm import tqdm

model.eval()

all_hashes = []

with torch.no_grad():

    for batch in tqdm(train_loader):

        # move batch to device
        batch = batch.to(device)

        # forward pass
        embedding, hash_vector = model(batch)

        # convert to binary hash (0/1)
        binary_hash = torch.sign(hash_vector)
        binary_hash = (binary_hash > 0).int()

        # move to CPU and convert to numpy
        all_hashes.append(binary_hash.cpu().numpy())


# combine all batches
hash_vectors = np.vstack(all_hashes)

print("Hash matrix shape:", hash_vectors.shape)


# convert to uint8 (required for FAISS binary index)
hash_vectors = hash_vectors.astype(np.uint8)

print("Hash dtype:", hash_vectors.dtype)


# pack bits (64 bits → 8 bytes)
hash_vectors = np.packbits(hash_vectors, axis=1)

print("Packed hash shape:", hash_vectors.shape)


# save hashes so you don't recompute
np.save("satellite_hash_vectors.npy", hash_vectors)

print("Hash vectors saved successfully!")
```
