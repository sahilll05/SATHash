# Satellite Image Hashing — FIXED VERSION
## BigEarthNet-S2 · Deep Spectral-Spatial Hash Retrieval

### Bugs fixed from original:
1. **Hash collapse** — original `create_pairs` used ALL positive labels; model learned distance=0 for everything → same hash for every image. Replaced with **NT-Xent (SimCLR) loss** which uses in-batch negatives automatically.
2. **Broken quantization loss** — `(h.sign() - h)²` is non-differentiable. Fixed to `(|h| - 1)²`.
3. **No bit-balance loss** — nothing stopped all bits from being identical. Added `independence_loss`.
4. **Hash↔file misalignment** — hash generation used shuffled DataLoader but retrieval used `os.listdir` order. Fixed with an ordered dataset.
5. **Visualization bug** — query shown as `retrieved_images[0]` (first match), not the actual query.

```python
# ============================================================
# PHASE 1 — PATHS
# ============================================================
train_path = r"D:\Projects\others\image-hashing\dataset\big-earth-net\BigEarthNet-S2\train"
val_path   = r"D:\Projects\others\image-hashing\dataset\big-earth-net\BigEarthNet-S2\validation"
test_path  = r"D:\Projects\others\image-hashing\dataset\big-earth-net\BigEarthNet-S2\test"
metadata_path = r"D:\Projects\others\image-hashing\dataset\big-earth-net\metadata.parquet"
```

```python
# ============================================================
# PHASE 2 — IMPORTS
# ============================================================
import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import rasterio
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# ============================================================
# PHASE 3 — DATASET
# ============================================================
class BigEarthDataset(Dataset):
    """
    Loads BigEarthNet-S2 .tif patches.
    Returns both the image AND its filename so hash↔file alignment is preserved.
    """
    def __init__(self, folder, limit=5000):
        self.folder = folder
        # Sort for deterministic, reproducible ordering
        all_files = sorted([f for f in os.listdir(folder) if f.endswith(".tif")])
        self.files = all_files[:limit]
        print(f"Dataset: {len(self.files)} images from {folder}")

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        path = os.path.join(self.folder, self.files[idx])
        with rasterio.open(path) as src:
            img = src.read().astype(np.float32)  # (C, H, W)

        # Sentinel-2 reflectance normalization (range ~0–10000)
        img = img / 10000.0

        # Clip to [0, 1] — removes anomalous high values caused by clouds/noise
        img = np.clip(img, 0.0, 1.0)

        return torch.tensor(img), self.files[idx]  # return filename too


train_dataset = BigEarthDataset(train_path, limit=5000)

# Training loader — shuffle is fine here, we only use it for training
train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True,
    num_workers=0,
    pin_memory=(device.type == "cuda")
)

# Ordered loader — used for hash generation so index i corresponds to train_dataset.files[i]
ordered_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=False,   # <-- CRITICAL: no shuffle so hashes align with file list
    num_workers=0
)

print("Ordered loader batches:", len(ordered_loader))
```

```python
# ============================================================
# PHASE 4 — DATA AUGMENTATION
# ============================================================
# NOTE: torchvision transforms expect (C, H, W) float tensors.
#       BigEarthNet has 10 bands — transforms work on the tensor directly.

def augment_image(img: torch.Tensor) -> torch.Tensor:
    """Apply spatial augmentations to a single (C, H, W) tensor."""
    if torch.rand(1) > 0.5:
        img = torch.flip(img, dims=[2])   # horizontal flip
    if torch.rand(1) > 0.5:
        img = torch.flip(img, dims=[1])   # vertical flip
    k = torch.randint(0, 4, (1,)).item()  # 0/90/180/270 rotation
    img = torch.rot90(img, k, dims=[1, 2])
    # Spectral jitter: slight per-band scaling
    scale = 1.0 + (torch.rand(img.shape[0], 1, 1) - 0.5) * 0.1  # ±5%
    img = torch.clamp(img * scale, 0.0, 1.0)
    return img
```

```python
# ============================================================
# PHASE 5 — MODEL ARCHITECTURE
# ============================================================
class SpectralHashNet(nn.Module):
    """
    Hybrid Spatial + Spectral encoder with a 64-bit deep hash layer.

    Architecture:
        Spatial branch  : ResNet-style 2D CNN (bands as channels)
        Spectral branch : 1D conv over the spectral dimension at each pixel
        Fusion          : concat → FC → embedding
        Hash layer      : Linear → Tanh → 64-bit code
    """
    def __init__(self, in_channels=10, embed_dim=256, hash_bits=64):
        super().__init__()

        # --- Spatial branch ---
        self.spatial = nn.Sequential(
            nn.Conv2d(in_channels, 32, 3, padding=1),
            nn.BatchNorm2d(32), nn.ReLU(),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64), nn.ReLU(),
            nn.MaxPool2d(2),               # 60×60

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128), nn.ReLU(),
            nn.MaxPool2d(2),               # 30×30

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256), nn.ReLU(),

            nn.AdaptiveAvgPool2d((1, 1))   # → (B, 256, 1, 1)
        )

        # --- Spectral branch (1D conv over bands, applied per-pixel) ---
        # We global-avg-pool spatial first, then learn spectral relationships
        self.spectral = nn.Sequential(
            nn.AdaptiveAvgPool2d((4, 4)),  # spatial downscale to 4×4
            nn.Flatten(),                  # B × (10*16) = B×160
            nn.Linear(in_channels * 16, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU()
        )

        # --- Fusion ---
        self.fusion = nn.Sequential(
            nn.Linear(256 + 128, embed_dim),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(embed_dim, embed_dim)
        )

        # --- Hash layer ---
        self.hash_layer = nn.Sequential(
            nn.Linear(embed_dim, embed_dim // 2),
            nn.ReLU(),
            nn.Linear(embed_dim // 2, hash_bits),
            nn.Tanh()   # outputs in (-1, 1); sign() → binary {-1, +1}
        )

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.ones_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        # x: (B, 10, H, W)
        sp_feat = self.spatial(x).squeeze(-1).squeeze(-1)    # (B, 256)
        se_feat = self.spectral(x)                            # (B, 128)

        fused = torch.cat([sp_feat, se_feat], dim=1)          # (B, 384)
        embedding = self.fusion(fused)                        # (B, embed_dim)
        embedding = F.normalize(embedding, dim=1)             # unit sphere

        hash_code = self.hash_layer(embedding)                # (B, hash_bits) in (-1,1)

        return embedding, hash_code


model = SpectralHashNet(in_channels=10, embed_dim=256, hash_bits=64).to(device)
print(model)
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"\nTrainable parameters: {total_params:,}")
```

```python
# ============================================================
# PHASE 6 — LOSS FUNCTIONS  (all 3 bugs fixed here)
# ============================================================

# -------------------------------------------------------
# FIX 1: NT-Xent Loss (SimCLR style)
# -------------------------------------------------------
# Why: Original contrastive loss used label=1 for EVERY pair
# (all augmented views treated as positive). With no negatives,
# the model minimises distance between ALL pairs → collapse.
#
# NT-Xent automatically uses every OTHER sample in the batch
# as a negative, so the model must discriminate between images.

def nt_xent_loss(z1: torch.Tensor, z2: torch.Tensor, temperature: float = 0.07) -> torch.Tensor:
    """
    NT-Xent (Normalized Temperature-scaled Cross Entropy) — SimCLR loss.
    z1, z2: L2-normalised embeddings of two views of the same batch, shape (N, D).
    For each sample i, positive = its augmented twin; negatives = all other 2N-2 samples.
    """
    N = z1.size(0)
    z = torch.cat([z1, z2], dim=0)                # (2N, D)
    z = F.normalize(z, dim=1)
    sim = torch.mm(z, z.T) / temperature           # (2N, 2N)

    # Remove self-similarity (diagonal)
    mask = torch.eye(2 * N, dtype=torch.bool, device=z.device)
    sim = sim.masked_fill(mask, float('-inf'))

    # Positive indices: for i in [0,N), positive is i+N; for i in [N,2N), positive is i-N
    labels = torch.cat([torch.arange(N, 2*N), torch.arange(N)]).to(device)

    return F.cross_entropy(sim, labels)


# -------------------------------------------------------
# FIX 2: Quantization Loss
# -------------------------------------------------------
# Original bug: (h.sign() - h)²  — .sign() has zero gradient almost everywhere,
# so this loss provides no useful gradient signal to the network.
#
# Fix: (|h| - 1)²  pushes tanh outputs toward ±1 (near binary) while being
# fully differentiable throughout the training range.

def quantization_loss(h: torch.Tensor) -> torch.Tensor:
    """Push each hash bit toward +1 or -1 (fully differentiable)."""
    return torch.mean((torch.abs(h) - 1.0) ** 2)


# -------------------------------------------------------
# FIX 3: Bit-Balance / Independence Loss
# -------------------------------------------------------
# Without this, all 64 bits can collapse to the same value.
# This loss:
#   (a) balance:  each bit should fire ~50% positive/negative in the batch
#   (b) decorr:   bits should not be correlated (encourages diverse codes)

def independence_loss(h: torch.Tensor) -> torch.Tensor:
    """
    h: (B, hash_bits) in (-1, 1)
    Returns a scalar loss that penalises:
      - Mean bit activation far from 0  (balance)
      - High correlation between bits   (independence)
    """
    B, K = h.shape

    # Balance: mean of each bit across the batch should be ~0
    balance = torch.mean(h, dim=0)           # (K,)
    balance_loss = torch.mean(balance ** 2)  # scalar

    # Independence: correlation matrix should be close to identity
    h_centered = h - h.mean(dim=0, keepdim=True)
    cov = torch.mm(h_centered.T, h_centered) / (B - 1)  # (K, K)
    # Penalise off-diagonal entries
    eye = torch.eye(K, device=h.device)
    decorr_loss = torch.mean((cov * (1 - eye)) ** 2)

    return balance_loss + 0.5 * decorr_loss


print("Loss functions defined.")
```

```python
# ============================================================
# PHASE 7 — TRAINING
# ============================================================
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)

# Cosine annealing gives a smooth lr decay — better than StepLR for this problem
epochs = 20
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-5)

# Loss weights
LAMBDA_Q   = 0.1   # quantization loss weight
LAMBDA_IND = 0.05  # independence loss weight
TEMPERATURE = 0.07 # NT-Xent temperature

print(f"Training for {epochs} epochs | batch_size=64 | LR=1e-3")
print(f"Loss = NT-Xent  +  {LAMBDA_Q}*quant  +  {LAMBDA_IND}*independence")
```

```python
history = {"total": [], "ntxent": [], "quant": [], "indep": []}

for epoch in range(epochs):
    model.train()
    epoch_ntxent = epoch_quant = epoch_indep = epoch_total = 0.0

    progress = tqdm(train_loader, desc=f"Epoch {epoch+1:02d}/{epochs}")

    for imgs, _ in progress:          # _ = filenames, not needed during training
        imgs = imgs.to(device)        # (B, 10, H, W)

        # Create two augmented views of every image in the batch
        view1 = torch.stack([augment_image(img) for img in imgs]).to(device)
        view2 = torch.stack([augment_image(img) for img in imgs]).to(device)

        emb1, hash1 = model(view1)
        emb2, hash2 = model(view2)

        # NT-Xent pulls same-image pairs together, pushes different-image pairs apart
        loss_ntxent = nt_xent_loss(emb1, emb2, temperature=TEMPERATURE)

        # Quantization: push bits toward ±1
        loss_quant = quantization_loss(hash1) + quantization_loss(hash2)

        # Independence: prevent bit collapse
        loss_indep = independence_loss(hash1) + independence_loss(hash2)

        loss = loss_ntxent + LAMBDA_Q * loss_quant + LAMBDA_IND * loss_indep

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # stability
        optimizer.step()

        epoch_ntxent += loss_ntxent.item()
        epoch_quant  += loss_quant.item()
        epoch_indep  += loss_indep.item()
        epoch_total  += loss.item()

        progress.set_postfix(loss=f"{loss.item():.4f}",
                             ntx=f"{loss_ntxent.item():.3f}")

    scheduler.step()

    history["total"].append(epoch_total)
    history["ntxent"].append(epoch_ntxent)
    history["quant"].append(epoch_quant)
    history["indep"].append(epoch_indep)

    print(f"Epoch {epoch+1:02d} | Total: {epoch_total:.3f} | "
          f"NT-Xent: {epoch_ntxent:.3f} | Quant: {epoch_quant:.3f} | Indep: {epoch_indep:.3f}")

print("\nTraining complete.")
```

```python
# Plot training curves
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].plot(history["total"], label="Total Loss", linewidth=2)
axes[0].plot(history["ntxent"], label="NT-Xent", linestyle="--")
axes[0].set_title("Training Loss")
axes[0].set_xlabel("Epoch")
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history["quant"], label="Quantization", color="orange")
axes[1].plot(history["indep"], label="Independence", color="green")
axes[1].set_title("Regularization Losses")
axes[1].set_xlabel("Epoch")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig("training_curves.png", dpi=100)
plt.show()
```

```python
# ============================================================
# PHASE 9 — HASH GENERATION  (FIX 4: use ordered_loader)
# ============================================================
# CRITICAL FIX: Use `ordered_loader` (shuffle=False), NOT `train_loader` (shuffle=True).
# Without this, hash[i] would not correspond to train_dataset.files[i].

model.eval()

all_hashes = []
all_files_ordered = []   # ground-truth aligned file list

with torch.no_grad():
    for imgs, fnames in tqdm(ordered_loader, desc="Generating hashes"):
        imgs = imgs.to(device)

        _, hash_code = model(imgs)               # (B, 64) in (-1, 1)

        # Binarise: +1 → 1, -1 → 0
        binary = (hash_code > 0).cpu().numpy().astype(np.uint8)  # (B, 64)
        all_hashes.append(binary)
        all_files_ordered.extend(fnames)         # keep track of order

hash_matrix = np.vstack(all_hashes)             # (N, 64)
print(f"Hash matrix shape: {hash_matrix.shape}")

# Sanity check — how many UNIQUE hashes did the model produce?
n_unique = len(np.unique(hash_matrix, axis=0))
print(f"Unique hashes: {n_unique} / {len(hash_matrix)}  "
      f"({'GOOD — diverse' if n_unique > 100 else 'BAD — collapsed, re-train!'})"
)

# Average Hamming diversity (higher is better, max=64)
sample_idx = np.random.choice(len(hash_matrix), min(500, len(hash_matrix)), replace=False)
sample = hash_matrix[sample_idx]
hamming_dists = []
for i in range(0, len(sample)-1, 1):
    for j in range(i+1, min(i+10, len(sample))):
        hamming_dists.append(np.sum(sample[i] != sample[j]))
print(f"Mean Hamming distance between random pairs: {np.mean(hamming_dists):.1f} / 64")
```

```python
# Pack 64 bits → 8 bytes for FAISS binary index
hash_packed = np.packbits(hash_matrix, axis=1)       # (N, 8)
print("Packed hash shape:", hash_packed.shape, "dtype:", hash_packed.dtype)

# Save both forms
np.save("satellite_hash_vectors.npy", hash_packed)
np.save("satellite_hash_matrix.npy", hash_matrix)    # unpacked, for analysis

# Save aligned file list — MUST stay in sync with hash rows
import json
with open("satellite_image_files.json", "w") as f:
    json.dump(all_files_ordered, f)

print(f"Saved {len(all_files_ordered)} hash↔file pairs")
```

```python
# ============================================================
# PHASE 10 — BUILD FAISS INDEX
# ============================================================
import faiss

# Binary flat index — Hamming distance search
index = faiss.IndexBinaryFlat(64)     # 64 bits
index.add(hash_packed)
print("FAISS index total vectors:", index.ntotal)

# Use the ordered file list from generation — guaranteed aligned
image_files = all_files_ordered
print("File list length:", len(image_files))
```

```python
# ============================================================
# PHASE 11 — METADATA LABELS
# ============================================================
import pandas as pd

metadata = pd.read_parquet(metadata_path)
print("Metadata shape:", metadata.shape)
print(metadata.head(2))
```

```python
label_map = {
    1:"Urban fabric", 2:"Industrial area", 3:"Arable land",
    4:"Permanent crops", 5:"Pastures", 6:"Complex cultivation",
    7:"Agriculture + natural vegetation", 8:"Agro-forestry",
    9:"Broad-leaved forest", 10:"Coniferous forest",
    11:"Mixed forest", 12:"Natural grassland", 13:"Moors / heathland",
    14:"Sclerophyllous vegetation", 15:"Transitional woodland",
    16:"Beaches / dunes", 17:"Inland wetlands", 18:"Coastal wetlands",
    19:"Inland waters", 20:"Marine waters"
}

image_labels = {}
for _, row in metadata.iterrows():
    image_id = row["patch_id"]
    labels = row["labels"]
    image_labels[image_id] = [label_map[l] for l in labels if l in label_map]

print("Label dictionary created:", len(image_labels), "entries")
```

```python
# ============================================================
# PHASE 12 — RETRIEVAL FUNCTION
# ============================================================
def retrieve_similar(query_image_path: str, k: int = 5):
    """
    Given a .tif file path, generate its hash and return the k most similar
    images from the FAISS index.

    Returns: (query_img_np, [(filename, hamming_dist), ...])
    """
    with rasterio.open(query_image_path) as src:
        img = src.read().astype(np.float32)   # (C, H, W)

    img_np = np.transpose(img, (1, 2, 0))     # (H, W, C) for display later

    # Normalise
    img = img / 10000.0
    img = np.clip(img, 0.0, 1.0)

    tensor = torch.tensor(img).unsqueeze(0).to(device)  # (1, C, H, W)

    model.eval()
    with torch.no_grad():
        _, hash_code = model(tensor)
        binary = (hash_code > 0).cpu().numpy().astype(np.uint8)  # (1, 64)
        packed = np.packbits(binary, axis=1)                     # (1, 8)

    D, I = index.search(packed, k + 1)     # +1 because query itself may be in index

    query_fname = os.path.basename(query_image_path)
    results = []
    for dist, idx in zip(D[0], I[0]):
        fname = image_files[idx]
        if fname != query_fname:           # skip exact self-match
            results.append((fname, int(dist)))
        if len(results) == k:
            break

    return img_np, results


print("Retrieval function ready.")
```

```python
# ============================================================
# PHASE 13 — TEST RETRIEVAL + VISUALISATION  (FIX 5: correct query display)
# ============================================================
query_index = 10
query_path  = os.path.join(train_path, image_files[query_index])

query_img, matches = retrieve_similar(query_path, k=4)

print(f"Query:  {image_files[query_index]}")
query_name = image_files[query_index]
print(f"Labels: {image_labels.get(query_name, ['(not in metadata)'])}")
print()
for fname, dist in matches:
    print(f"  Match: {fname}  |  Hamming={dist}  |  Labels: {image_labels.get(fname, ['?'])}")
```

```python
# Load retrieved images
retrieved_imgs = []
for fname, dist in matches:
    path = os.path.join(train_path, fname)
    with rasterio.open(path) as src:
        img = src.read()
    retrieved_imgs.append((np.transpose(img, (1, 2, 0)), dist))


def to_rgb(img_hwc: np.ndarray) -> np.ndarray:
    """Extract bands 3,2,1 (Red, Green, Blue) and normalise to [0,1] for display."""
    rgb = img_hwc[:, :, [3, 2, 1]].astype(np.float32)
    # Per-image 2nd–98th percentile stretch
    p2, p98 = np.percentile(rgb, 2), np.percentile(rgb, 98)
    rgb = np.clip((rgb - p2) / (p98 - p2 + 1e-6), 0, 1)
    return rgb


# -------------------------------------------------------
# FIX 5: Show actual query image in first panel
# Original bug: retrieved_images[0] (which was a MATCH) was shown as "Query"
# -------------------------------------------------------
n = 1 + len(retrieved_imgs)
fig, axes = plt.subplots(1, n, figsize=(4 * n, 4))

# --- Query panel ---
axes[0].imshow(to_rgb(query_img))
axes[0].set_title("Query", fontsize=12, fontweight="bold")
axes[0].axis("off")

# --- Match panels ---
for i, (img, dist) in enumerate(retrieved_imgs):
    axes[i + 1].imshow(to_rgb(img))
    axes[i + 1].set_title(f"Match {i+1}\nHamming={dist}", fontsize=10)
    axes[i + 1].axis("off")

plt.suptitle("Satellite Image Retrieval (BigEarthNet-S2)", fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig("retrieval_results.png", dpi=120, bbox_inches="tight")
plt.show()
```

```python
# ============================================================
# PHASE 14 — EVALUATION: HASH QUALITY METRICS
# ============================================================

# 1. Unique hash ratio (was 1/5000 before → 100% collapse)
n_unique = len(np.unique(hash_matrix, axis=0))
print(f"Unique hashes: {n_unique} / {len(hash_matrix)} ({100*n_unique/len(hash_matrix):.1f}%)")

# 2. Bit activation rate (each bit should fire ~50% of the time)
bit_rates = hash_matrix.mean(axis=0)        # (64,)
balanced = np.sum((bit_rates > 0.3) & (bit_rates < 0.7))
print(f"Balanced bits (30–70% activation): {balanced} / {hash_matrix.shape[1]}")
print(f"Bit activation rates — min: {bit_rates.min():.3f}, max: {bit_rates.max():.3f}, mean: {bit_rates.mean():.3f}")

# 3. Mean pairwise Hamming distance
idx = np.random.choice(len(hash_matrix), 200, replace=False)
sample = hash_matrix[idx]
dists = []
for i in range(len(sample)):
    for j in range(i+1, min(i+20, len(sample))):
        dists.append(np.sum(sample[i] != sample[j]))
print(f"Mean pairwise Hamming distance: {np.mean(dists):.1f} (ideal ~32 for random codes)")

# 4. Plot bit activation distribution
plt.figure(figsize=(12, 3))
plt.bar(range(len(bit_rates)), bit_rates, color="steelblue", alpha=0.7)
plt.axhline(0.5, color="red", linestyle="--", label="ideal 50%")
plt.axhline(0.3, color="orange", linestyle=":", alpha=0.7)
plt.axhline(0.7, color="orange", linestyle=":", alpha=0.7, label="30–70% band")
plt.xlabel("Bit index")
plt.ylabel("Activation rate")
plt.title("Per-bit Activation Rate Across Training Set")
plt.legend()
plt.tight_layout()
plt.savefig("bit_activation.png", dpi=100)
plt.show()
```

```python
# ============================================================
# PHASE 15 — LABEL PRECISION@K EVALUATION
# ============================================================
# For each test image, retrieve k=5 similar images and compute
# the fraction that share at least one label with the query.

def precision_at_k(query_fname, k=5) -> float:
    """Returns fraction of top-k results sharing at least one label with query."""
    query_path_full = os.path.join(train_path, query_fname)
    _, results = retrieve_similar(query_path_full, k=k)

    query_labels = set(image_labels.get(query_fname, []))
    if not query_labels:
        return None  # skip if no labels

    hits = 0
    for fname, _ in results:
        match_labels = set(image_labels.get(fname, []))
        if query_labels & match_labels:  # intersection
            hits += 1
    return hits / k


# Evaluate on 100 random images
eval_indices = np.random.choice(len(image_files), 100, replace=False)
precisions = []

for idx in tqdm(eval_indices, desc="Evaluating P@5"):
    fname = image_files[idx]
    p = precision_at_k(fname, k=5)
    if p is not None:
        precisions.append(p)

print(f"\nPrecision@5 over {len(precisions)} images: {np.mean(precisions):.3f} "
      f"(std={np.std(precisions):.3f})")
```

```python
# ============================================================
# PHASE 16 — DEMO: MULTI-QUERY GRID
# ============================================================
# Show 4 different queries and their top-3 matches in a grid.

demo_indices = [5, 50, 200, 500]
K_DEMO = 3

fig, axes = plt.subplots(len(demo_indices), K_DEMO + 1, figsize=(4 * (K_DEMO + 1), 4 * len(demo_indices)))

for row, qi in enumerate(demo_indices):
    fname  = image_files[qi]
    qpath  = os.path.join(train_path, fname)
    qimg, matches_ = retrieve_similar(qpath, k=K_DEMO)

    qlabels = image_labels.get(fname, ["?"])
    axes[row][0].imshow(to_rgb(qimg))
    axes[row][0].set_title(f"Query\n{', '.join(qlabels[:2])}", fontsize=8)
    axes[row][0].axis("off")

    for col, (mfname, dist) in enumerate(matches_):
        mpath = os.path.join(train_path, mfname)
        with rasterio.open(mpath) as src:
            mimg = np.transpose(src.read(), (1, 2, 0))
        mlabels = image_labels.get(mfname, ["?"])
        axes[row][col + 1].imshow(to_rgb(mimg))
        axes[row][col + 1].set_title(f"Match {col+1} | H={dist}\n{', '.join(mlabels[:2])}", fontsize=8)
        axes[row][col + 1].axis("off")

plt.suptitle("Multi-Query Retrieval Demo", fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig("multi_query_demo.png", dpi=120, bbox_inches="tight")
plt.show()
```
