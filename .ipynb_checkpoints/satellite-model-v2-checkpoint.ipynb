# Satellite Image Hashing — v2 (Collapse Fixed)
## Changes from v1:
- **Much stronger independence/quantization loss weights** to break bit collapse
- **Warmup scheduler**: let NT-Xent stabilise before forcing binarization
- **Metadata ID fix**: strip `.tif` and match patch_id correctly
- **How to know if working**: inline diagnostics after every 5 epochs
- **More epochs** (30) + **smaller batch** (32) for more diverse negatives

```python
# ============================================================
# PHASE 1 — PATHS
# ============================================================
train_path    = r"D:\Projects\others\image-hashing\dataset\big-earth-net\BigEarthNet-S2\train"
val_path      = r"D:\Projects\others\image-hashing\dataset\big-earth-net\BigEarthNet-S2\validation"
test_path     = r"D:\Projects\others\image-hashing\dataset\big-earth-net\BigEarthNet-S2\test"
metadata_path = r"D:\Projects\others\image-hashing\dataset\big-earth-net\metadata.parquet"
```

```python
# ============================================================
# PHASE 2 — IMPORTS
# ============================================================
import os, json, math
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import rasterio
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# ============================================================
# PHASE 3 — DATASET
# ============================================================
class BigEarthDataset(Dataset):
    def __init__(self, folder, limit=5000):
        self.folder = folder
        all_files = sorted([f for f in os.listdir(folder) if f.endswith(".tif")])
        self.files = all_files[:limit]
        print(f"Dataset: {len(self.files)} images")

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        path = os.path.join(self.folder, self.files[idx])
        with rasterio.open(path) as src:
            img = src.read().astype(np.float32)   # (C, H, W)
        img = img / 10000.0
        img = np.clip(img, 0.0, 1.0)
        return torch.tensor(img), self.files[idx]


train_dataset = BigEarthDataset(train_path, limit=5000)

# Smaller batch = more diverse negatives per epoch for NT-Xent
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True,  num_workers=0)
ordered_loader = DataLoader(train_dataset, batch_size=64, shuffle=False, num_workers=0)
```

```python
# ============================================================
# PHASE 4 — AUGMENTATION
# ============================================================
def augment_image(img: torch.Tensor) -> torch.Tensor:
    if torch.rand(1) > 0.5:
        img = torch.flip(img, dims=[2])
    if torch.rand(1) > 0.5:
        img = torch.flip(img, dims=[1])
    k = torch.randint(0, 4, (1,)).item()
    img = torch.rot90(img, k, dims=[1, 2])
    # Spectral jitter ±10%
    scale = 1.0 + (torch.rand(img.shape[0], 1, 1) - 0.5) * 0.2
    img = torch.clamp(img * scale, 0.0, 1.0)
    return img
```

```python
# ============================================================
# PHASE 5 — MODEL
# ============================================================
class SpectralHashNet(nn.Module):
    def __init__(self, in_channels=10, embed_dim=256, hash_bits=64):
        super().__init__()

        self.spatial = nn.Sequential(
            nn.Conv2d(in_channels, 32, 3, padding=1),
            nn.BatchNorm2d(32), nn.ReLU(),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256), nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1))
        )

        self.spectral = nn.Sequential(
            nn.AdaptiveAvgPool2d((4, 4)),
            nn.Flatten(),
            nn.Linear(in_channels * 16, 128), nn.ReLU(),
            nn.Linear(128, 128), nn.ReLU()
        )

        self.fusion = nn.Sequential(
            nn.Linear(256 + 128, embed_dim), nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(embed_dim, embed_dim)
        )

        # Hash layer — NO Tanh here; we apply it only during hash generation
        # During training, keep raw logits for better gradient flow
        self.hash_layer = nn.Sequential(
            nn.Linear(embed_dim, embed_dim // 2), nn.ReLU(),
            nn.Linear(embed_dim // 2, hash_bits)
            # Tanh applied externally so we can scale it during training
        )

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.Linear):
                nn.init.xavier_normal_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.ones_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        sp = self.spatial(x).squeeze(-1).squeeze(-1)   # (B, 256)
        se = self.spectral(x)                           # (B, 128)
        fused = torch.cat([sp, se], dim=1)              # (B, 384)
        embedding = F.normalize(self.fusion(fused), dim=1)
        hash_logits = self.hash_layer(embedding)        # (B, 64) — raw, no Tanh yet
        return embedding, hash_logits


model = SpectralHashNet(in_channels=10, embed_dim=256, hash_bits=64).to(device)
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Parameters: {total_params:,}")
```

```python
# ============================================================
# PHASE 6 — LOSS FUNCTIONS
# ============================================================

def nt_xent_loss(z1, z2, temperature=0.07):
    """SimCLR NT-Xent — uses in-batch negatives, prevents collapse."""
    N = z1.size(0)
    z = F.normalize(torch.cat([z1, z2], dim=0), dim=1)  # (2N, D)
    sim = torch.mm(z, z.T) / temperature
    mask = torch.eye(2 * N, dtype=torch.bool, device=z.device)
    sim = sim.masked_fill(mask, float('-inf'))
    labels = torch.cat([torch.arange(N, 2*N), torch.arange(N)]).to(device)
    return F.cross_entropy(sim, labels)


def quantization_loss(h_logits):
    """
    Push each logit to have |tanh(x)| ≈ 1.
    Equivalent to penalising values in the 'soft' middle zone near 0.
    """
    h = torch.tanh(h_logits)
    return torch.mean((torch.abs(h) - 1.0) ** 2)


def independence_loss(h_logits):
    """
    Two sub-losses:
      1. Balance: each bit should activate ~50% of time (mean → 0)
      2. Decorrelation: bits should not be correlated with each other
    """
    h = torch.tanh(h_logits)  # (B, K)
    B, K = h.shape

    # Balance
    balance_loss = torch.mean(h.mean(dim=0) ** 2)

    # Decorrelation
    h_c = h - h.mean(dim=0, keepdim=True)
    cov = torch.mm(h_c.T, h_c) / (B - 1)  # (K, K)
    eye = torch.eye(K, device=h.device)
    decorr_loss = torch.mean((cov * (1 - eye)) ** 2)

    return balance_loss + decorr_loss


print("Loss functions ready.")
```

```python
# ============================================================
# PHASE 7 — TRAINING  (key fix: much stronger reg weights + warmup)
# ============================================================

# ── Why the previous run still collapsed ──────────────────────────────
# Your independence loss was ~65 and quant loss was ~16 at epoch 20.
# With LAMBDA_IND=0.05 and LAMBDA_Q=0.1, their weighted contribution was:
#   0.05 × 65 = 3.25  vs  NT-Xent ≈ 10 → the reg signal was too weak.
# We need to increase these until the bit activation chart shows spread.
# ──────────────────────────────────────────────────────────────────────

EPOCHS      = 30
WARMUP      = 5      # epochs before turning on reg losses
TEMPERATURE = 0.07
LAMBDA_Q    = 1.0    # was 0.1 — increased 10× to actually penalise non-binary
LAMBDA_IND  = 0.5    # was 0.05 — increased 10× to break bit correlation

optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS, eta_min=1e-5)

history = {"total": [], "ntxent": [], "quant": [], "indep": [], "unique_pct": []}

print(f"Training {EPOCHS} epochs | batch=32 | LR=1e-3")
print(f"Warmup {WARMUP} epochs (only NT-Xent), then add reg losses")
print(f"LAMBDA_Q={LAMBDA_Q}  LAMBDA_IND={LAMBDA_IND}")
print()
```

```python
for epoch in range(EPOCHS):
    model.train()
    e_ntxent = e_quant = e_indep = e_total = 0.0

    # Gradually ramp up reg losses after warmup (avoid destabilising early training)
    reg_scale = min(1.0, (epoch - WARMUP + 1) / 5) if epoch >= WARMUP else 0.0

    progress = tqdm(train_loader, desc=f"Epoch {epoch+1:02d}/{EPOCHS}")

    for imgs, _ in progress:
        imgs = imgs.to(device)
        v1 = torch.stack([augment_image(img) for img in imgs]).to(device)
        v2 = torch.stack([augment_image(img) for img in imgs]).to(device)

        emb1, h1 = model(v1)
        emb2, h2 = model(v2)

        loss_ntx = nt_xent_loss(emb1, emb2, TEMPERATURE)
        loss_q   = quantization_loss(h1) + quantization_loss(h2)
        loss_ind = independence_loss(h1) + independence_loss(h2)

        loss = loss_ntx + reg_scale * (LAMBDA_Q * loss_q + LAMBDA_IND * loss_ind)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()

        e_ntxent += loss_ntx.item()
        e_quant  += loss_q.item()
        e_indep  += loss_ind.item()
        e_total  += loss.item()

        progress.set_postfix(total=f"{loss.item():.3f}",
                             ntx=f"{loss_ntx.item():.3f}",
                             reg=f"{reg_scale:.2f}")

    scheduler.step()

    history["total"].append(e_total)
    history["ntxent"].append(e_ntxent)
    history["quant"].append(e_quant)
    history["indep"].append(e_indep)

    # ── INLINE HEALTH CHECK every 5 epochs ────────────────────────────
    # This is how you know whether the model is working:
    if (epoch + 1) % 5 == 0 or epoch == EPOCHS - 1:
        model.eval()
        sample_hashes = []
        with torch.no_grad():
            for imgs_s, _ in ordered_loader:
                _, h = model(imgs_s.to(device))
                binary = (torch.tanh(h) > 0).cpu().numpy().astype(np.uint8)
                sample_hashes.append(binary)
                if len(sample_hashes) * 64 >= 500:  # quick sample
                    break
        sample_hashes = np.vstack(sample_hashes)[:500]
        n_unique = len(np.unique(sample_hashes, axis=0))
        bit_rates = sample_hashes.mean(axis=0)
        balanced = int(np.sum((bit_rates > 0.3) & (bit_rates < 0.7)))
        unique_pct = 100 * n_unique / len(sample_hashes)
        history["unique_pct"].append(unique_pct)

        status = "✅ GOOD" if n_unique > 200 else ("⚠️  PARTIAL" if n_unique > 50 else "❌ COLLAPSED")
        print(f"  [{status}] Unique hashes: {n_unique}/500 ({unique_pct:.1f}%) | "
              f"Balanced bits: {balanced}/64 | "
              f"Bit mean: {bit_rates.mean():.3f}")
        model.train()
    # ──────────────────────────────────────────────────────────────────

    print(f"Epoch {epoch+1:02d} | Total:{e_total:.2f} | "
          f"NT-Xent:{e_ntxent:.2f} | Q:{e_quant:.2f} | Ind:{e_indep:.2f} | reg_scale:{reg_scale:.2f}")

print("\nTraining complete.")
```

```python
# ============================================================
# HOW TO READ THE TRAINING HEALTH CHECK
# ============================================================
# After every 5 epochs you see a line like:
#
#   [✅ GOOD]    Unique hashes: 412/500 (82.4%) | Balanced bits: 48/64 | Bit mean: 0.501
#   [⚠️  PARTIAL] Unique hashes: 87/500 (17.4%) | Balanced bits: 21/64 | Bit mean: 0.62
#   [❌ COLLAPSED] Unique hashes: 3/500  (0.6%)  | Balanced bits:  1/64 | Bit mean: 0.99
#
# Targets to aim for by epoch 30:
#   Unique hashes  > 60%  (ideally >80%)
#   Balanced bits  > 40 / 64
#   Bit mean       close to 0.5
#   Hamming dist   > 20 / 64 (checked in Phase 9)
#
# If still collapsed after epoch 10:
#   → Increase LAMBDA_IND to 1.0 or 2.0 and re-run
#   → Or increase LAMBDA_Q to 2.0
print("(No code to run — documentation cell)")
```

```python
# ============================================================
# PHASE 9 — HASH GENERATION
# ============================================================
model.eval()
all_hashes = []
all_files_ordered = []

with torch.no_grad():
    for imgs, fnames in tqdm(ordered_loader, desc="Generating hashes"):
        _, h = model(imgs.to(device))
        binary = (torch.tanh(h) > 0).cpu().numpy().astype(np.uint8)
        all_hashes.append(binary)
        all_files_ordered.extend(fnames)

hash_matrix = np.vstack(all_hashes)
print(f"Hash matrix: {hash_matrix.shape}")

# ── FULL QUALITY REPORT ───────────────────────────────────────────────
n_unique = len(np.unique(hash_matrix, axis=0))
bit_rates = hash_matrix.mean(axis=0)
balanced  = int(np.sum((bit_rates > 0.3) & (bit_rates < 0.7)))

# Hamming distances
idx = np.random.choice(len(hash_matrix), 300, replace=False)
dists = []
for i in range(len(idx)):
    for j in range(i+1, min(i+15, len(idx))):
        dists.append(int(np.sum(hash_matrix[idx[i]] != hash_matrix[idx[j]])))

print()
print("═" * 55)
print("  HASH QUALITY REPORT")
print("═" * 55)
print(f"  Unique hashes   : {n_unique} / {len(hash_matrix)} "
      f"({100*n_unique/len(hash_matrix):.1f}%)  [target: >60%]")
print(f"  Balanced bits   : {balanced} / 64              [target: >40]")
print(f"  Bit act. mean   : {bit_rates.mean():.3f}                  [target: ~0.5]")
print(f"  Mean Hamming    : {np.mean(dists):.1f} / 64              [target: >20]")
print(f"  Min  Hamming    : {np.min(dists)}")
print(f"  Max  Hamming    : {np.max(dists)}")
print("═" * 55)

if n_unique / len(hash_matrix) < 0.1:
    print("\n❌ Still collapsed — increase LAMBDA_IND to 1.0 and re-run")
elif n_unique / len(hash_matrix) < 0.6:
    print("\n⚠️  Partial — model is learning but not fully spread")
    print("   Try: LAMBDA_IND=1.0, LAMBDA_Q=2.0, more epochs")
else:
    print("\n✅ Model is working well — proceed to retrieval")
```

```python
# ============================================================
# PHASE 10 — FAISS INDEX
# ============================================================
import faiss
index = faiss.IndexBinaryFlat(64)
index.add(hash_packed)
image_files = all_files_ordered
print("FAISS index:", index.ntotal, "vectors")
```

```python
# ============================================================
# PHASE 11 — METADATA  (FIX: handle string labels from parquet)
# ============================================================
import pandas as pd

metadata = pd.read_parquet(metadata_path)
print("Metadata shape:", metadata.shape)
print("Columns:", list(metadata.columns))
print()
# Check what the labels column actually looks like
print("Sample patch_id:", metadata["patch_id"].iloc[0])
print("Sample labels:  ", metadata["labels"].iloc[0])
print("Labels type:    ", type(metadata["labels"].iloc[0]))
```

```python
# ── FIX: BigEarthNet parquet labels are already strings, not int codes ──
# The metadata patch_id does NOT have .tif extension; filenames do.
# We strip .tif from filenames when looking up labels.

image_labels = {}

for _, row in metadata.iterrows():
    patch_id = row["patch_id"]   # e.g. "S2A_MSIL2A_20170803T094031_N9999_R036_T34TCR_32_09"
    labels   = row["labels"]     # already a list of strings in BigEarthNet parquet

    # Handle both list-of-strings and list-of-ints
    if labels and isinstance(labels[0], str):
        readable = list(labels)
    else:
        label_map = {
            1:"Urban fabric", 2:"Industrial", 3:"Arable land",
            4:"Permanent crops", 5:"Pastures", 6:"Complex cultivation",
            7:"Agriculture+NatVeg", 8:"Agro-forestry",
            9:"Broad-leaved forest", 10:"Coniferous forest",
            11:"Mixed forest", 12:"Natural grassland",
            13:"Moors/heathland", 14:"Sclerophyllous",
            15:"Transitional woodland", 16:"Beaches/dunes",
            17:"Inland wetlands", 18:"Coastal wetlands",
            19:"Inland waters", 20:"Marine waters"
        }
        readable = [label_map.get(l, str(l)) for l in labels]

    image_labels[patch_id] = readable

print(f"Label dictionary: {len(image_labels)} entries")

# Verify lookup works by testing a filename
test_fname = image_files[0]          # e.g. "S2A_MSIL2A_..._32_09.tif"
test_id    = test_fname.replace(".tif", "")
print(f"Test lookup: '{test_id}' → {image_labels.get(test_id, 'NOT FOUND')}")
```

```python
def get_labels(fname):
    """Look up labels for a filename, stripping .tif automatically."""
    key = fname.replace(".tif", "")
    return image_labels.get(key, ["(no label)"])
```

```python
# ============================================================
# PHASE 12 — RETRIEVAL
# ============================================================
def retrieve_similar(query_image_path, k=5):
    with rasterio.open(query_image_path) as src:
        img = src.read().astype(np.float32)
    img_np = np.transpose(img, (1, 2, 0))

    img = np.clip(img / 10000.0, 0, 1)
    tensor = torch.tensor(img).unsqueeze(0).to(device)

    model.eval()
    with torch.no_grad():
        _, h = model(tensor)
        binary = (torch.tanh(h) > 0).cpu().numpy().astype(np.uint8)
        packed = np.packbits(binary, axis=1)

    D, I = index.search(packed, k + 1)
    query_fname = os.path.basename(query_image_path)
    results = []
    for dist, idx_r in zip(D[0], I[0]):
        fname = image_files[idx_r]
        if fname != query_fname:
            results.append((fname, int(dist)))
        if len(results) == k:
            break
    return img_np, results


print("Retrieval function ready.")
```

```python
# ============================================================
# PHASE 13 — TEST RETRIEVAL
# ============================================================
query_index = 10
query_path  = os.path.join(train_path, image_files[query_index])
query_img, matches = retrieve_similar(query_path, k=4)

qname = image_files[query_index]
print(f"Query:  {qname}")
print(f"Labels: {get_labels(qname)}")
print()
for fname, dist in matches:
    label_match = bool(set(get_labels(qname)) & set(get_labels(fname)))
    match_icon  = "✅" if label_match else "❌"
    print(f"  {match_icon} {fname}  Hamming={dist}  Labels: {get_labels(fname)}")
```

```python
def to_rgb(img_hwc):
    rgb = img_hwc[:, :, [3, 2, 1]].astype(np.float32)
    p2, p98 = np.percentile(rgb, 2), np.percentile(rgb, 98)
    return np.clip((rgb - p2) / (p98 - p2 + 1e-6), 0, 1)


retrieved_imgs = []
for fname, dist in matches:
    with rasterio.open(os.path.join(train_path, fname)) as src:
        img = np.transpose(src.read(), (1, 2, 0))
    retrieved_imgs.append((img, dist))

n = 1 + len(retrieved_imgs)
fig, axes = plt.subplots(1, n, figsize=(4 * n, 4))

axes[0].imshow(to_rgb(query_img))
axes[0].set_title(f"Query\n{', '.join(get_labels(qname)[:2])}", fontsize=9, fontweight="bold")
axes[0].axis("off")

for i, (img, dist) in enumerate(retrieved_imgs):
    fname_i = matches[i][0]
    label_match = bool(set(get_labels(qname)) & set(get_labels(fname_i)))
    border_col = "green" if label_match else "red"
    axes[i+1].imshow(to_rgb(img))
    for sp in axes[i+1].spines.values():
        sp.set_edgecolor(border_col); sp.set_linewidth(3)
    axes[i+1].set_title(f"Match {i+1} H={dist}\n{', '.join(get_labels(fname_i)[:2])}",
                        fontsize=8)
    axes[i+1].axis("off")

plt.suptitle("Green border = label match  |  Red = mismatch", fontsize=11)
plt.tight_layout()
plt.savefig("retrieval_v2.png", dpi=120, bbox_inches="tight")
plt.show()
```

```python
# ============================================================
# PHASE 14 — PRECISION@5 EVALUATION  (fixed metadata lookup)
# ============================================================

def precision_at_k(fname, k=5):
    qpath = os.path.join(train_path, fname)
    _, results = retrieve_similar(qpath, k=k)
    q_labels = set(get_labels(fname))
    if not q_labels or q_labels == {"(no label)"}:
        return None
    hits = sum(1 for mfname, _ in results if set(get_labels(mfname)) & q_labels)
    return hits / k


eval_indices = np.random.choice(len(image_files), 100, replace=False)
precisions = []

for idx in tqdm(eval_indices, desc="Evaluating P@5"):
    p = precision_at_k(image_files[idx], k=5)
    if p is not None:
        precisions.append(p)

if precisions:
    print(f"\nPrecision@5: {np.mean(precisions):.3f}  (std={np.std(precisions):.3f})")
    print(f"Evaluated on {len(precisions)} images")
    print()
    print("Interpretation:")
    print("  > 0.6  → Good retrieval quality")
    print("  0.3–0.6 → Acceptable")
    print("  < 0.3  → Model needs more training or stronger reg losses")
else:
    print("No valid labels found — check metadata path and patch_id format above")
```

```python
# ============================================================
# PHASE 15 — MULTI-QUERY DEMO GRID
# ============================================================
demo_indices = [5, 50, 200, 500]
K_DEMO = 3

fig, axes = plt.subplots(len(demo_indices), K_DEMO + 1,
                         figsize=(4 * (K_DEMO + 1), 4 * len(demo_indices)))

for row, qi in enumerate(demo_indices):
    fname = image_files[qi]
    qimg, matches_ = retrieve_similar(os.path.join(train_path, fname), k=K_DEMO)
    qlabels = get_labels(fname)

    axes[row][0].imshow(to_rgb(qimg))
    axes[row][0].set_title(f"Query\n{', '.join(qlabels[:2])}", fontsize=8)
    axes[row][0].axis("off")

    for col, (mfname, dist) in enumerate(matches_):
        with rasterio.open(os.path.join(train_path, mfname)) as src:
            mimg = np.transpose(src.read(), (1, 2, 0))
        mlabels = get_labels(mfname)
        hit = bool(set(qlabels) & set(mlabels))
        axes[row][col+1].imshow(to_rgb(mimg))
        for sp in axes[row][col+1].spines.values():
            sp.set_edgecolor("green" if hit else "red"); sp.set_linewidth(3)
        axes[row][col+1].set_title(f"Match {col+1} H={dist}\n{', '.join(mlabels[:2])}",
                                   fontsize=8)
        axes[row][col+1].axis("off")

plt.suptitle("Multi-Query Demo  |  Green=label match  Red=mismatch", fontsize=12, y=1.01)
plt.tight_layout()
plt.savefig("multi_query_demo_v2.png", dpi=120, bbox_inches="tight")
plt.show()
```

```python
# ============================================================
# PHASE 16 — SUMMARY SCORECARD
# ============================================================
print("━" * 60)
print("  FINAL MODEL SCORECARD")
print("━" * 60)

n_u = len(np.unique(hash_matrix, axis=0))
br  = hash_matrix.mean(axis=0)
bal = int(np.sum((br > 0.3) & (br < 0.7)))

def grade(val, good, ok):
    return "✅" if val >= good else ("⚠️ " if val >= ok else "❌")

unique_pct = 100 * n_u / len(hash_matrix)
hd_mean    = np.mean(dists) if dists else 0
p5         = np.mean(precisions) if precisions else 0

print(f"  {grade(unique_pct, 60, 20)} Unique hashes : {unique_pct:.1f}%   (good≥60, ok≥20)")
print(f"  {grade(bal, 40, 20)}  Balanced bits : {bal}/64      (good≥40, ok≥20)")
print(f"  {grade(hd_mean, 20, 10)} Mean Hamming  : {hd_mean:.1f}/64   (good≥20, ok≥10)")
print(f"  {grade(p5, 0.6, 0.3)}  Precision@5   : {p5:.3f}      (good≥0.6, ok≥0.3)")
print("━" * 60)
```
